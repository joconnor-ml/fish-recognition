```python
import json
import pandas as pd
import glob
boxes = []
for f in glob.glob("/home/alpha/annotations/*.json"):
    print(f)
    j = json.load(open(f))
    for element in j:
        try:
            box = element["annotations"][0]
        except:
            box = {}
        box["class"] = element["class"]
        box["filename"] = element["filename"].split("/")[-1]
        boxes.append(box)
```

```python
df = pd.DataFrame(boxes)
df = df.set_index("filename")
```
